```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
sns.set() # Sets the default dark grid seaborns style, even for pandas plots!
```

```python
df = pd.read_csv('datasets/heart.csv')

df.info()
```

```python
df.describe()
```

```python
# Correlation Heatmap
corr = df.corr()

mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

plt.figure(figsize=(14, 8))

sns.heatmap(corr, mask=mask, annot=False,cmap='RdYlGn', linewidths=0.2, annot_kws={'size':20})
plt.show()
```

```python
np.abs(corr['target']).sort_values(ascending=False).iloc[1:].plot.bar()
plt.xlabel('Feature')
plt.ylabel('Correlation')
plt.title('Abs Correlation with Disease', fontsize=18)
```
