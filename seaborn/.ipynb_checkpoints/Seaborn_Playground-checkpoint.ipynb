```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set() # Sets the default dark grid seaborns style, even for pandas plots!
```

```python
for p in sns.palettes.SEABORN_PALETTES:
    print(p)
    sns.set_palette(p)
    sns.palplot(sns.color_palette())
    plt.show()
```
