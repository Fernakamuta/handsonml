# 3) Batch Gradient

```python
# To furthers comparison, let's record theta in the thetas_dict

thetas_path_dict = {
    'batch': [],
    'stochastic': [],
    'mini': []
}
```

```python
def add_x0(X):
    """Add x0 = 1 to each instance."""
    shape = X.shape
    return np.c_[np.ones(shape), X]

def calc_gradient_vectors(X, y, theta, m=None):
    """Calculate the gradient vector for some theta parameters."""
    if not m:
        m = len(X)
    return 2/m * X.T.dot(X.dot(theta) - y)

def predict(X, theta):
    """Makes a predict on some X vectors."""
    return X.dot(theta)
```

```python
eta = 0.1
n_iterations = 1000
# m = 100
theta = np.random.randn(2,1)

thetas_path_dict['batch'] = []

for iteration in range(n_iterations):
    if iteration % 50 == 0:
        print(f'i {iteration: } ->  theta {theta.flatten()}')
    gradients = calc_gradient_vectors(X_b, y, theta)
    theta = theta - eta * gradients # Nice!
    thetas_path_dict['batch'].append(theta)
```

```python
def plot_current_model_line(theta, X_new=np.array([[0],[2]])):
    """Plots the linear regression model. X_new is some random points"""
    y_new = predict(add_x0(X_new), theta)
    plt.plot(X_new, y_new, 'g--')
```

```python
def plot_gradients(X, y, theta=np.random.randn(2,1), eta=0.1, iterations=100):
    X_b = add_x0(X)
    
    plt.plot(X, y, 'b.')
    plot_current_model_line(theta)
    
    for iteration in range(iterations):
        gradients = calc_gradient_vectors(X_b, y, theta)
        theta = theta - eta * gradients
        plot_current_model_line(theta)
    
plot_gradients(X, y)
```

```python
np.random.seed(42)
theta = np.random.randn(2,1)  # random initialization

plt.figure(figsize=(20,12))
plt.subplot(131)
plot_gradients(X, y, theta, eta=0.02)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.xlabel("$x_1$", rotation=0, fontsize=18)


plt.subplot(132)
plot_gradients(X, y,theta, eta=0.1)
plt.title('Eta 0.1')

plt.subplot(133)
plot_gradients(X, y, theta, eta=0.5)
plt.title('Eta 0.4')
```

# 4) Stochastic Gradient Descent
The main problem with batch gradient descent is taking the whole X dataset on every gradient calculation.
For large datasets, this may be a problem!
Stochastic gradient can be a nice alternative, since it computes the gradient using only one randomly chosen instance.

```python
theta_path_sgd = []
m = len(X_b)
np.random.seed(42)
```

```python
thetas_path_dict['stochastic'] = []
n_epochs = 50
t0, t1 = 5, 50

def learning_schedule(t, t0=5, t1=50):
    return t0 / (t + t1)

theta = np.random.randn(2, 1)

plt.plot(X, y, '.')
plt.xlabel('$x_1$', fontsize=18)
plt.ylabel('$y$', fontsize=18, rotation=0)

# plot_current_model_line(theta)

for epoch in range(n_epochs):
    for i in range(m):
        if epoch == 0 or i == 0:
            # Plotting the first epoch and every first iteration from every epoch 
            plot_current_model_line(theta)

        random_index = np.random.randint(m)
        
        xi = X_b[random_index:random_index+1]
        yi = y[random_index:random_index+1]
        
        gradients = calc_gradient_vectors(xi, yi, theta)

        eta = learning_schedule(epoch * m + i)
        theta = theta - eta * gradients
        thetas_path_dict['stochastic'].append(theta)
        
print('Final theta:', theta)
```

# 5) Mini-Batch Gradient Descent
The mini-batch gradient is a mix of batch and stochastic gradient descent.
Gradient vectors are computed using small random samples called mini-batches.
As a result, Mini-Batch GD may walk closer to the global minimum than SGD.
But as a tradeoff, if the cost function is very irregular, it is more likely to be stuck in a local minimum or plateau.

```python
thetas_path_dict['mini'] = []
n_epochs = 50
t0, t1 = 5, 50

def learning_schedule(t, t0=5, t1=50):
    return t0 / (t + t1)

theta = np.random.randn(2, 1)

plt.plot(X, y, '.')
plt.xlabel('$x_1$', fontsize=18)
plt.ylabel('$y$', fontsize=18, rotation=0)

# plot_current_model_line(theta)

for epoch in range(n_epochs):
    for i in range(m):
        if epoch == 0 or i == 0:
            # Plotting the first epoch and every first iteration from every epoch 
            plot_current_model_line(theta)

        random_index = np.random.randint(m)
        
        xi = X_b[random_index:random_index+1]
        yi = y[random_index:random_index+1]
        
        gradients = calc_gradient_vectors(xi, yi, theta)

        eta = learning_schedule(epoch * m + i)
        theta = theta - eta * gradients
        thetas_path_dict['stochastic'].append(theta)
        
print('Final theta:', theta)
```

```python
thetas_path_dict['mini'] = []

n_iterations = 50
minibatch_size = 20

np.random.seed(42)
theta = np.random.randn(2,1)  # random initialization

t0, t1 = 200, 1000
def learning_schedule(t):
    return t0 / (t + t1)

t = 0
for epoch in range(n_iterations):
    shuffled_indices = np.random.permutation(m)
    X_b_shuffled = X_b[shuffled_indices]
    y_shuffled = y[shuffled_indices]
    for i in range(0, m, minibatch_size):
        t += 1
        xi = X_b_shuffled[i:i+minibatch_size]
        yi = y_shuffled[i:i+minibatch_size]
        gradients = calc_gradient_vectors(xi, yi, theta)
        eta = learning_schedule(t)
        theta = theta - eta * gradients
        thetas_path_dict['mini'].append(theta)

print('Final theta:', theta)
```

```python
theta_path_bgd = np.array(thetas_path_dict['batch'])
theta_path_sgd = np.array(thetas_path_dict['stochastic'])
theta_path_mgd = np.array(thetas_path_dict['mini'])
```

```python
plt.figure(figsize=(12,12))
plt.plot(theta_path_sgd[:, 0], theta_path_sgd[:, 1], "r-s", linewidth=1, label="Stochastic")
plt.plot(theta_path_mgd[:, 0], theta_path_mgd[:, 1], "g-+", linewidth=2, label="Mini-batch")
plt.plot(theta_path_bgd[:, 0], theta_path_bgd[:, 1], "b-o", linewidth=3, label="Batch")
plt.legend(loc="upper left", fontsize=16)
plt.xlabel(r"$\theta_0$", fontsize=20)
plt.ylabel(r"$\theta_1$   ", fontsize=20, rotation=0)
plt.axis([2.5, 4.5, 2.3, 3.9])
plt.show()
```

# Learning Curves
How can you tell if your model is overfitting or underfitting the data?

In chapter 2, we did it with cross-validation.
If the model performs well on the training but performs poorly in the cross-validation metrics, then it's overfitting.
If the model performs poorly on both, then it's underfitting. This is one way to tell when a model is too simple or too complex

Another way is to look at the **Learning Curves**

```python
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

def plot_learning_curves(model, X, y):
    print("Plotting Curves ohyeah")
    X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=10)
    
    train_errors = []
    val_errors = []
    
    m = len(X_train)
    
    for i in range(1, m):
        X_i, y_i = X_train[:i], y[:i]
        model.fit(X_i, y_i)
        
        y_i_pred = model.predict(X_i)
        train_errors.append(mean_squared_error(y_i, y_i_pred))
        
        y_i_val_pred = model.predict(X_val)
        val_errors.append(mean_squared_error(y_i_val_pred, y_val))       
    
    plt.plot(np.sqrt(train_errors), 'r-+', label="train")
    plt.plot(np.sqrt(val_errors), '-', label='val')
    plt.xlabel("Training size", fontsize=14)
    plt.ylabel("RMSE", fontsize=14)
```

```python
plot_learning_curves(LinearRegression(), X, y)
# plt.axis([0, 80, 0, 3])                         # not shown in the book
plt.show()                                      # not shown
```

# Regularized Linear Models

Regularizing your model is a good way to avoid your model overfitting

Ridge, Lasso and ElasticNet are three ways to constrain the weights

```python
from sklearn.linear_model import Ridge

np.random.seed(42)
m = 20
X = 3 * np.random.rand(m, 1)
y = 1 + 0.5 * X + np.random.randn(m, 1) / 1.5
X_new = np.linspace(0, 3, 100).reshape(100, 1)

def plot_model(model_class, polynomial, alphas, **model_kargs):
    for alpha, style in zip(alphas, ("b-", "g--", "r:")):
        model = model_class(alpha, **model_kargs) if alpha > 0 else LinearRegression()
        if polynomial:
            model = Pipeline([
                    ("poly_features", PolynomialFeatures(degree=10, include_bias=False)),
                    ("std_scaler", StandardScaler()),
                    ("regul_reg", model),
                ])
        model.fit(X, y)
        y_new_regul = model.predict(X_new)
        lw = 2 if alpha > 0 else 1
        plt.plot(X_new, y_new_regul, style, linewidth=lw, label=r"$\alpha = {}$".format(alpha))
    plt.plot(X, y, "b.", linewidth=3)
    plt.legend(loc="upper left", fontsize=15)
    plt.xlabel("$x_1$", fontsize=18)
    plt.axis([0, 3, 0, 4])

plt.figure(figsize=(8,4))
plt.subplot(121)
plot_model(Ridge, polynomial=False, alphas=(0, 10, 100), random_state=42)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.subplot(122)
plot_model(Ridge, polynomial=True, alphas=(0, 10**-5, 1), random_state=42)

plt.show()
```

```python
np.random.seed(42)
m = 100
X = 6 * np.random.rand(m, 1) - 3
y = 2 + X + 0.5 * X**2 + np.random.randn(m, 1)

X_train, X_val, y_train, y_val = train_test_split(X[:50], y[:50].ravel(), test_size=0.5, random_state=10)

poly_scaler = Pipeline([
        ("poly_features", PolynomialFeatures(degree=90, include_bias=False)),
        ("std_scaler", StandardScaler()),
    ])

X_train_poly_scaled = poly_scaler.fit_transform(X_train)
X_val_poly_scaled = poly_scaler.transform(X_val)

sgd_reg = SGDRegressor(max_iter=1,
                       tol=-np.infty,
                       penalty=None,
                       eta0=0.0005,
                       warm_start=True,
                       learning_rate="constant",
                       random_state=42)

n_epochs = 500
train_errors, val_errors = [], []
for epoch in range(n_epochs):
    sgd_reg.fit(X_train_poly_scaled, y_train)
    y_train_predict = sgd_reg.predict(X_train_poly_scaled)
    y_val_predict = sgd_reg.predict(X_val_poly_scaled)
    train_errors.append(mean_squared_error(y_train, y_train_predict))
    val_errors.append(mean_squared_error(y_val, y_val_predict))

best_epoch = np.argmin(val_errors)
best_val_rmse = np.sqrt(val_errors[best_epoch])

plt.annotate('Best model',
             xy=(best_epoch, best_val_rmse),
             xytext=(best_epoch, best_val_rmse + 1),
             ha="center",
             arrowprops=dict(facecolor='green', shrink=0.05),
             fontsize=16,
            )

best_val_rmse -= 0.03  # just to make the graph look better
plt.plot([0, n_epochs], [best_val_rmse, best_val_rmse], "k:", linewidth=2)
plt.plot(np.sqrt(val_errors), "b-", linewidth=3, label="Validation set")
plt.plot(np.sqrt(train_errors), "r--", linewidth=2, label="Training set")
plt.legend(loc="upper right", fontsize=14)
plt.xlabel("Epoch", fontsize=14)
plt.ylabel("RMSE", fontsize=14)
plt.show()
```

```python
from sklearn.base import clone
sgd_reg = SGDRegressor(max_iter=1, tol=-np.infty, warm_start=True, penalty=None,
                       learning_rate="constant", eta0=0.0005, random_state=42)

minimum_val_error = float("inf")
best_epoch = None
best_model = None
for epoch in range(1000):
    sgd_reg.fit(X_train_poly_scaled, y_train)  # continues where it left off
    y_val_predict = sgd_reg.predict(X_val_poly_scaled)
    val_error = mean_squared_error(y_val, y_val_predict)
    if val_error < minimum_val_error:
        minimum_val_error = val_error
        best_epoch = epoch
        best_model = clone(sgd_reg)
```

```python
t1a, t1b, t2a, t2b = -1, 3, -1.5, 1.5

# ignoring bias term
t1s = np.linspace(t1a, t1b, 500)
t2s = np.linspace(t2a, t2b, 500)
t1, t2 = np.meshgrid(t1s, t2s)
T = np.c_[t1.ravel(), t2.ravel()]
Xr = np.array([[-1, 1], [-0.3, -1], [1, 0.1]])
yr = 2 * Xr[:, :1] + 0.5 * Xr[:, 1:]

J = (1/len(Xr) * np.sum((T.dot(Xr.T) - yr.T)**2, axis=1)).reshape(t1.shape)

N1 = np.linalg.norm(T, ord=1, axis=1).reshape(t1.shape)
N2 = np.linalg.norm(T, ord=2, axis=1).reshape(t1.shape)

t_min_idx = np.unravel_index(np.argmin(J), J.shape)
t1_min, t2_min = t1[t_min_idx], t2[t_min_idx]

t_init = np.array([[0.25], [-1]])
```

```python
def bgd_path(theta, X, y, l1, l2, core = 1, eta = 0.1, n_iterations = 50):
    path = [theta]
    for iteration in range(n_iterations):
        gradients = core * 2/len(X) * X.T.dot(X.dot(theta) - y) + l1 * np.sign(theta) + 2 * l2 * theta

        theta = theta - eta * gradients
        path.append(theta)
    return np.array(path)

plt.figure(figsize=(12, 8))
for i, N, l1, l2, title in ((0, N1, 0.5, 0, "Lasso"), (1, N2, 0,  0.1, "Ridge")):
    JR = J + l1 * N1 + l2 * N2**2
    
    tr_min_idx = np.unravel_index(np.argmin(JR), JR.shape)
    t1r_min, t2r_min = t1[tr_min_idx], t2[tr_min_idx]

    levelsJ=(np.exp(np.linspace(0, 1, 20)) - 1) * (np.max(J) - np.min(J)) + np.min(J)
    levelsJR=(np.exp(np.linspace(0, 1, 20)) - 1) * (np.max(JR) - np.min(JR)) + np.min(JR)
    levelsN=np.linspace(0, np.max(N), 10)
    
    path_J = bgd_path(t_init, Xr, yr, l1=0, l2=0)
    path_JR = bgd_path(t_init, Xr, yr, l1, l2)
    path_N = bgd_path(t_init, Xr, yr, np.sign(l1)/3, np.sign(l2), core=0)

    plt.subplot(221 + i * 2)
    plt.grid(True)
    plt.axhline(y=0, color='k')
    plt.axvline(x=0, color='k')
    plt.contourf(t1, t2, J, levels=levelsJ, alpha=0.9)
    plt.contour(t1, t2, N, levels=levelsN)
    plt.plot(path_J[:, 0], path_J[:, 1], "w-o")
    plt.plot(path_N[:, 0], path_N[:, 1], "y-^")
    plt.plot(t1_min, t2_min, "rs")
    plt.title(r"$\ell_{}$ penalty".format(i + 1), fontsize=16)
    plt.axis([t1a, t1b, t2a, t2b])
    if i == 1:
        plt.xlabel(r"$\theta_1$", fontsize=20)
    plt.ylabel(r"$\theta_2$", fontsize=20, rotation=0)

    plt.subplot(222 + i * 2)
    plt.grid(True)
    plt.axhline(y=0, color='k')
    plt.axvline(x=0, color='k')
    plt.contourf(t1, t2, JR, levels=levelsJR, alpha=0.9)
    plt.plot(path_JR[:, 0], path_JR[:, 1], "w-o")
    plt.plot(t1r_min, t2r_min, "rs")
    plt.title(title, fontsize=16)
    plt.axis([t1a, t1b, t2a, t2b])
    if i == 1:
        plt.xlabel(r"$\theta_1$", fontsize=20)

plt.show()
```

### 2) Suppose the features of your training set have different scales. What algorithms might suffer from this and how? What can you do about it?

Gradient Descent derivates will suffer from different scales.
The gradient path to the minimum will be very irregular (enlongated bowl shape), so it may require a larger number of iterations to converge to a minimum.

The normal equation or SVD approaches wont suffer from it.

Regularized models may find a suboptimal solution, since regularization penalize large weights, features with smaller values will tend to be ignored compared to larger values features.

### 3) Can Gradient Descent get stuck in a local minimum when training a linear model?

No, because linear models cost functions are convex!

### 4) Do all GD algorithms lead to the same model provided you let them run long enough? 

If the cost function is convex and the learning rate it not too high, them all models will find the optimal solution and end up with fairly similar models. However unless you decrease the learning rate, stochastic GD and mini-batch GD will never trully convert.
